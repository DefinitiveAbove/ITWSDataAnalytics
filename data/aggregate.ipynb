```python
from scipy.spatial import distance
import numpy as np

import cProfile
import csv

# https://codereview.stackexchange.com/questions/28207/finding-the-closest-point-to-a-list-of-points
def closest_node(node, nodes):
    closest_index = distance.cdist([node], nodes).argmin()
    return nodes[closest_index]

def normalize_dataset(csvFileName: str, baseDist):
    threshold = distance.euclidean((0, 0), baseDist) * 1.25
    datasetDict = dict()
    nodeList = []
    # read the csv into a python dictionary not with pd
    with open(csvFileName, 'r') as csvFile:
        csvReader = csv.DictReader(csvFile)
        fields = list(csvReader.fieldnames)
        fields.remove("lat")
        fields.remove("lon")
        for row in csvReader:
            lat = float(row["lat"])
            lon = float(row["lon"])
            nodeList.append((lat, lon))
            if lat not in datasetDict:
                datasetDict[lat] = dict()
            if lon not in datasetDict[lat]:
                datasetDict[lat][lon] = dict()
            for field in fields:
                datasetDict[lat][lon][field] = float(row[field])
    lat = float(-90)

    nodeList = np.asarray(nodeList)

    # pr = cProfile.Profile()
    for i in range(0, 180*2 + 1):
        # pr.enable()
        print(lat)
        lon = float(-180)
        for x in range(0, 360*2 + 1):
            if lat in datasetDict and lon in datasetDict[lat]:
                for field in fields:
                    finalDataset[lat][lon][field] = datasetDict[lat][lon][field]
            else:
                closest = closest_node((lat, lon), nodeList)
                dist = distance.euclidean((lat, lon), closest)
                if dist <= threshold:
                    for field in fields:
                        finalDataset[lat][lon][field] = datasetDict[closest[0]][closest[1]][field]
                else:
                    for field in fields:
                        finalDataset[lat][lon][field] = None
            lon = round(lon + 0.5, 1)
        lat = round(lat + 0.5, 1)
        # pr.disable()
        # import pstats, io
        # from pstats import SortKey
        # s = io.StringIO()
        # sortby = SortKey.CUMULATIVE
        # ps = pstats.Stats(pr, stream=s).sort_stats(sortby)
        # ps.print_stats()
        # print(s.getvalue())
        # return
    
finalDataset = dict()
lat = -90


for i in range(0, 180*2 + 1):
    finalDataset[lat] = dict()
    lon = -180
    for x in range(0, 360*2 + 1):
        finalDataset[lat][lon] = dict()
        lon = round(lon + 0.5, 1)
    lat = round(lat + 0.5, 1)

csvFiles = [("nasa_power/meterology/meterology.csv", (0.5, 0.625)), ("nasa_power/radiation/radiation.csv", (1, 1)), ("uea_cru/cloud_cover/cloud_cover.csv", (0.5, 0.5)), ("uea_cru/potential_evapotranspiration/potential_evapotranspiration.csv", (0.5, 0.5))]

for csvFile in csvFiles:
    normalize_dataset(csvFile[0], csvFile[1])


    
    
    # uniqueLat = dataframe.lat.unique()
    # uniqueLon = dataframe.lon.unique()

    # lat = float(-90)
    # for i in range(0, 180*2 + 1):
    #     lon = float(-180)
    #     for x in range(0, 360*2 + 1):
    #         finalDataset[lat][lon]

    #         lon = round(lon + 0.5, 1)
    #     lat = round(lat + 0.5, 1)
```

```python
# save finalDataset to a csv
with open("dataset.csv", 'w') as csvFile:
    datasetKeys = list(finalDataset[0][0].keys())
    
    csvFile.write("lat,lon," + ",".join(datasetKeys) + "\n")
    for lat in finalDataset:
        for lon in finalDataset[lat]:
            indivOutput = [lat, lon]
            for key in datasetKeys:
                indivOutput.append(finalDataset[lat][lon][key])
            csvFile.write(",".join([str(x) for x in indivOutput]) + "\n")

with open("dataset_filtered.csv", 'w') as csvFile:
    datasetKeys = list(finalDataset[0][0].keys())
    
    csvFile.write("lat,lon," + ",".join(datasetKeys) + "\n")
    for lat in finalDataset:
        for lon in finalDataset[lat]:
            indivOutput = [lat, lon]
            shouldWrite = True
            for key in datasetKeys:
                if finalDataset[lat][lon][key] is None:
                    shouldWrite = False
                    break
                indivOutput.append(finalDataset[lat][lon][key])
            if shouldWrite:
                csvFile.write(",".join([str(x) for x in indivOutput]) + "\n")
```
